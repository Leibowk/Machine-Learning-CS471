```python
def getFitNValid(X, y, fit):
    n,_ = X.shape
    rowsFit = math.floor(fit*n)
    return X[:rowsFit, :], X[rowsFit:, :], y[:rowsFit], y[rowsFit:]


def testHyperparam(n_estimat, xTrainS,yTrainS, xValidS, yValidS):
    clasf=RandomForestRegressor(n_estimators=n_estimat)
    clasf.fit(xTrainS,yTrainS)
    yPredS=clasf.predict(xValidS)
    errorS = abs(yPredS - yValidS)
    acurracy = 100 - (100 * np.mean(errorS / yValidS))
    meanAbsError = round(np.mean(errorS), 5)
    print("N currently is: ", n_estimat, ' Mean Absolute Error:', meanAbsError, 'degrees.')
    print("Acurracy: ", acurracy)
#     clasf=RandomForestRegressor(n_estimators=n_estimat)
#     clasf.fit(xTrain,yTrain)
#     yPred=clasf.predict(xValid)
#     error = abs(yPred - yValid)
    return meanAbsError
```

```python
X_train = np.loadtxt("X_train", delimiter = ",")
y_train = np.loadtxt("y_train", delimiter = ",")
X_test = np.loadtxt("X_test", delimiter = ",")
```

```python
#xGetError and yGetError for finding error rate.
xGetParams, xGetError, yGetParams, yGetError = getFitNValid(X_train, y_train, .9)

#xTrain and yTrain for training, xValid, yValid for hyperparamater selection
xTrain, xValid, yTrain, yValid = getFitNValid(xGetParams, yGetParams, .9)

#really small for testing
xTrainS, xValidS, yTrainS, yValidS = getFitNValid(xGetParams, yGetParams, .15)
```

```python
n_estimat = 100
clasf=RandomForestRegressor(n_estimators=n_estimat)
clasf.fit(xTrain,yTrain)
yPred=clasf.predict(xValid)
error = abs(yPred - yValid)

print('Mean Absolute Error:', round(np.mean(error), 2), 'degrees.')
```

```python
bestHyper = float('inf')
bestMeanAbs = float('inf') 
for n_estimat in range(1, 100):
    meanAbsError = testHyperparam(n_estimat, xTrainS,yTrainS, xValidS, yValidS)
    if meanAbsError < bestMeanAbs:
        bestMeanAbs = meanAbsError
        bestHyper = n_estimat
    
print("Final Hyper paramater was: ", bestHyper, " which gave a mean absolute error of: ", bestMeanAbs)
```

```python
bestHyper = float('inf')
bestMeanAbs = float('inf') 
n_estimat = 100
while n_estimat <= 1000:
    meanAbsError = testHyperparam(n_estimat, xTrainS,yTrainS, xValidS, yValidS)
    
    if meanAbsError < bestMeanAbs:
        bestMeanAbs = meanAbsError
        bestHyper = n_estimat
    n_estimat = n_estimat + 100
    
print("Final Hyper paramater was: ", bestHyper, " which gave a mean absolute error of: ", bestMeanAbs)
```
